```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data", 
                 header=None, names=['Class', 'Odor', 'Habitat'], usecols=[0, 5, 22], dtype="category")
df
```

Now that the data is ready, we will build our model and see how well the Odor column and the Habitat column fit the model. This score determines which column is a better predictor for determing the mushroom's class (edible or poisonous). Judging from the graphs produced last week, I expect the Odor class to be the better predictor.

```python
model = sklearn.linear_model.LogisticRegression()
response = df["Class"].cat.codes
for predictor in ["Odor", "Habitat"]:
    inputs = pd.get_dummies(df[predictor])
    model.fit(inputs, response)
    print(predictor, "Score:", model.score(inputs, response))
```

The Odor column had a higher score for fitting the model than the Habitat column, therfore it is a better predictor. In the future, I might want to use cross validation, however since there were eight inputs to eight thousand rows I didn't feel like overfitting would be an issue.
